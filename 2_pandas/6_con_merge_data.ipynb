#### Data Concatination and Merging -

```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
    'city' : ['mumbai', 'delhi', 'banglore'],
    'temperature' : [32, 45, 30],
    'humidity' : [80, 60, 78]
})
india_weather
```

```python
usa_weather = pd.DataFrame({
    'city' : ['new york', 'chicago', 'orlando'],
    'temperature' : [21, 14, 35],
    'humidity' : [68, 65, 75]
})
usa_weather
```

```python
df = pd.concat([india_weather, usa_weather], ignore_index=True)
df
```

```python
df = pd.concat([india_weather, usa_weather], keys=['india', 'usa'])
df
```

```python
df.loc['india']
```

```python
df.loc['usa']
```

```python
temperature_df = pd.DataFrame({
    'city' : ['mumbai', 'delhi', 'banglore'],
    'temperature' : [40, 32, 38]
},
index=[1, 2, 3]
)

temperature_df
```

```python
windspeed_df = pd.DataFrame({
    'city' : ['delhi', 'mumbai'],
    'windspeed' : [7, 12]
},

index=[2, 1])

windspeed_df
```

```python
df = pd.concat([temperature_df, windspeed_df], axis=1)
df
```

```python
df1 = pd.DataFrame({
    'city' : ['new york', 'chicago', 'orlando', 'baltimore'],
    'temperature' : [32, 45, 30, 25],
})
df1
```

```python
df2 = pd.DataFrame({
    'city' : ['new york', 'chicago', 'orlando', 'baltimore'],
    'humidity' : [68, 65, 75, 80]
})
df2
```

```python
df3 = pd.merge(df1, df2, on='city')
df3
```

```python
df1 = pd.DataFrame({
    'city' : ['new york', 'chicago', 'orlando', 'baltimore'],
    'temperature' : [32, 45, 30, 25],
})
df1
```

```python
df2 = pd.DataFrame({
    'city' : ['new york', 'chicago', 'orlando', 'san diego'],
    'humidity' : [68, 65, 75, 80]
})
df2
```

```python
df3 = pd.merge(df1, df2, on='city')
df3
```

```python
df3 = pd.merge(df1, df2, on='city', how='outer')
df3
```

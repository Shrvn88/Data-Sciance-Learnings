#### Handle Missing Data -

```python
import pandas as pd
```

```python
df = pd.read_csv("D:\Dec'24\DSCourse\data\weather_data.csv", parse_dates=['day'])
```

```python
df.head()
```

```python
type(df.day[0])
```

```python
df.set_index('day', inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.fillna(0)
```

```python
df
```

```python
df.fillna({
    'temperature' : df.temperature.mean(),
    'windspeed' : df.windspeed.mean(),
    'event' : 'no event'      #df.event.mode()[0]
})
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
df.fillna(method='bfill', axis='columns')
```

```python
df.fillna(method='ffill', limit=1)
```

```python
df.interpolate()
```

```python
df.dropna(thresh=2)
```

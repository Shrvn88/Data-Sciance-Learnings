```python
import pandas as pd

df_sales = pd.read_excel("D:\Dec'24\DSCourse\data\linechart.xlsx")

df_sales.head()
```

```python
from matplotlib import pyplot as plt

plt.plot(df_sales['Quarter'], df_sales['Fridge'], color='red', label='fridge')
plt.plot(df_sales['Quarter'], df_sales['Dishwasher'], color='green', label='dishwasher')
plt.plot(df_sales['Quarter'], df_sales['Washing Machine'], color='black', label='washing machine')

plt.legend()

plt.show()
```

```python
total_sales = df_sales[['Fridge', 'Dishwasher', 'Washing Machine']].sum()
total_sales
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct='%1i%%', explode=(0.1, 0, 0), shadow=True, startangle=120)

plt.show()
```

```python
df_sales.plot(kind='bar', x='Quarter')
plt.xticks(rotation=45)
plt.show()
```
